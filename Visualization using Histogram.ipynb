```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d= pd.read_csv("Iris.csv")
```

```python
d.hist(['PetalLengthCm'], bins = 20, color = "green") 
plt.title("Petal Length in cm") 
plt.xlabel("Petal_Length_cm") 
plt.ylabel("Count") 
```

```python
d.hist(['PetalWidthCm'], bins = 20, color = "green") 
plt.title("Petal Widt in cm") 
plt.xlabel("Petal_Widt_cm") 
plt.ylabel("Count") 
```

```python
d.hist(['SepalLengthCm'], bins = 20, color = "blue") 
plt.title("Sepal Length in cm") 
plt.xlabel("Sepal_Length_cm") 
plt.ylabel("Count") 
```

```python
d.hist(['SepalWidthCm'], bins = 20, color = "blue") 
plt.title("Sepal Width in cm") 
plt.xlabel("Sepal_Width_cm") 
plt.ylabel("Count") 
```
